```python
import ast

def convert_ast(obj):
  try:
    data = ast.literal_eval(obj)
    return [i['name'] for i in data if isinstance(i, dict) and 'name' in i]
  except:
    return []
```

```python
movies['genres']=movies['genres'].apply(convert_ast)
```

```python
movies['keywords']=movies['keywords'].apply(convert_ast)
```

```python
def get_top_cast(obj):
  try:
    data = ast.literal_eval(obj)
    return [i['name'] for i in data[:3]]  # top 3 actors
  except:
    return []
```

```python
movies['cast']=movies['cast'].apply(get_top_cast)
movies['cast']
```

```python
movies.head(1)
```

```python
def get_director(obj):
  try:
    data = ast.literal_eval(obj)
    return [i['name'] for i in data if i.get('job') == 'Director']
  except:
    return []
```

```python
movies['crew']=movies['crew'].apply(get_director)
```

```python
# function use garera bichko gap or spaces hatayeko
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'].apply(lambda x: [x]) + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x)).apply(lambda x: x.lower())
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(5)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=movies['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df[new_df['title']=='Avatar']
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
